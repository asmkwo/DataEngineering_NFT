## Question 5

Compter le nombre moyen de vues en fonction de la catégorie.

```python
categories_mean_views = list(collection.aggregate([{"$group" : {"_id" : "$category_id", "Nombre moyen de vues" : {"$avg" : "$views"}}}]))
```

## Question 6 

Récupérer les chaines Youtube avec la plus grande moyenne de likes.

```python
top_channels_likes = list(collection.aggregate([{"$group" : {"_id" : "$channel_title", 
                                                             "Moyenne de likes" : {"$avg" : "$likes"}}},
                                               {"$sort" : {"Moyenne de likes" : -1}}]))
top_channels_likes
```
